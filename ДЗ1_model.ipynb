# **Строим логистическую регрессию - угадываем уровень дохода**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('adult.csv', sep = ',')
data.head()
```

```python
data.info()
```

```python
data['race'].unique()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# попробуем выбрать какие-нибудь признаки
selectedColumns = data[ [ 'age', 'workclass', 'fnlwgt','education','educational-num','marital-status','occupation','relationship','race','gender','capital-gain','capital-loss','hours-per-week','native-country' ] ]
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'workclass' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'education' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'marital-status' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'occupation' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'relationship' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'race' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'gender' ] )
selectedColumns = pd.get_dummies( selectedColumns, columns = [ 'native-country' ] )

X=selectedColumns
X.head()
```

```python
# целевая переменная (столбец income)  является категориальной
# переведем значения столбца в числа, оставив один столбец

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit( data['income'] )
```

```python
# записываем в переменную y преобразованный столбец income

y = pd.Series( data = le.transform( data['income'] ) )
y.head()
```

```python
model = LogisticRegression(max_iter=100000)
```

```python
X
```

```python
y.unique()
```

```python
# обучаем модель

model.fit( X, y )
predictions = model.predict( X )
predictions[0:15]

```

```python
model.score(X, y)
```
